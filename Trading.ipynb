```python
import yfinance as yf
import pandas as pd

# 1. Define the ticker symbol (e.g., Apple)
ticker_symbol = "AAPL"

# 2. Download the last 1 year of data
print(f"Downloading data for {ticker_symbol}...")
data = yf.download(ticker_symbol, period="1y", interval="1d")

# 3. Display the first 5 rows to confirm it worked
print("\nSuccess! Here are the first 5 rows of your data:")
print(data.head())
```

```python
import pandas_ta as ta
import pandas as pd # Import pandas for MultiIndex check

# Ensure we are using the 'data' variable from the previous step

# Flatten the MultiIndex columns if they exist
# yfinance can return a MultiIndex, which pandas_ta does not handle directly
if isinstance(data.columns, pd.MultiIndex):
    # This creates a single-level index, taking the second level for 'Price' columns
    # and the first level for 'Volume' column.
    data.columns = [col[1] if col[0] == 'Price' else col[0] for col in data.columns]

# 1. Calculate RSI (Length 14 is standard)
data['RSI'] = data.ta.rsi(length=14)

# 2. Calculate MACD (Fast=12, Slow=26, Signal=9 are standard)
# This appends the MACD columns directly to 'data'
data.ta.macd(fast=12, slow=26, signal=9, append=True)

# 3. Calculate Bollinger Bands (Length 20, Std Dev 2)
# This appends the Bollinger Bands columns directly to 'data'
data.ta.bbands(length=20, std=2, append=True)

# 4. Clean up: Drop rows with NaN values (the first few rows won't have calculations)
data.dropna(inplace=True)

# 5. Check the result
print("Success! Indicators added. Here are the last 5 rows:")
print(data[['Close', 'RSI', 'MACD_12_26_9', 'BBL_20_2.0_2.0', 'BBU_20_2.0_2.0']].tail())
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# 1. Create a figure with 2 subplots (Price on top, RSI on bottom)
fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.1,
    subplot_titles=(f'{ticker_symbol} Price & Bollinger Bands', 'RSI'),
    row_width=[0.3, 0.7] # Top chart is taller (70%), Bottom chart is smaller (30%)
)

# 2. Add Candlestick Chart (Open, High, Low, Close)
fig.add_trace(go.Candlestick(
    x=data.index,
    open=data['Open'], high=data['High'],
    low=data['Low'], close=data['Close'],
    name='OHLC'
), row=1, col=1)

# 3. Add Bollinger Bands (Upper and Lower)
# Upper Band
fig.add_trace(go.Scatter(
    x=data.index, y=data['BBU_20_2.0_2.0'],
    line=dict(color='gray', width=1),
    name='Upper BB'
), row=1, col=1)

# Lower Band
fig.add_trace(go.Scatter(
    x=data.index, y=data['BBL_20_2.0_2.0'],
    line=dict(color='gray', width=1),
    name='Lower BB'
), row=1, col=1)

# 4. Add RSI Trace (Bottom Panel)
fig.add_trace(go.Scatter(
    x=data.index, y=data['RSI'],
    line=dict(color='purple', width=2),
    name='RSI'
), row=2, col=1)

# 5. Add Overbought (70) and Oversold (30) lines for RSI
fig.add_hline(y=70, line_dash='dash', line_color='red', annotation_text='Overbought', row=2, col=1)
fig.add_hline(y=30, line_dash='dash', line_color='green', annotation_text='Oversold', row=2, col=1)

# 6. Customize Layout
fig.update_layout(
    title_text=f'{ticker_symbol} Stock Analysis with RSI and Bollinger Bands',
    xaxis_rangeslider_visible=False, # Hide the range slider for a cleaner look
    height=700,
    template='plotly_dark'
)

# 7. Show the plot
fig.show()
```

```python
import numpy as np

# 1. Create a copy of the data to avoid messing up the original
backtest_df = data.copy()

# 2. Define the Signal (0 = Hold, 1 = Buy, -1 = Sell)
backtest_df['Signal'] = 0

# 3. Apply the Strategy Logic
# If RSI < 30, set Signal to 1 (Long/Buy)
backtest_df.loc[backtest_df['RSI'] < 30, 'Signal'] = 1

# If RSI > 70, set Signal to -1 (Short/Sell)
backtest_df.loc[backtest_df['RSI'] > 70, 'Signal'] = -1

# 4. Handle "Holding" positions
# If signal is 0, we keep the previous position (forward fill)
backtest_df['Position'] = backtest_df['Signal'].replace(to_replace=0, method='ffill')

# 5. Calculate Returns
# Market Return: The daily % change of the stock price
backtest_df['Market_Return'] = backtest_df['Close'].pct_change()

# Strategy Return: Our position (from yesterday) * Today's Market Return
# We shift(1) because we make the trade based on yesterday's RSI close
backtest_df['Strategy_Return'] = backtest_df['Position'].shift(1) * backtest_df['Market_Return']

# 6. Calculate Portfolio Value (Starting with $10,000)
initial_capital = 10000
backtest_df['Portfolio_Value'] = initial_capital * (1 + backtest_df['Strategy_Return']).cumprod()

# 7. Print Results
final_portfolio_value = backtest_df['Portfolio_Value'].iloc[-1]
total_return_pct = ((final_portfolio_value / initial_capital) - 1) * 100

print(f"Initial Investment: ${initial_capital}")
print(f"Final Portfolio Value: ${final_portfolio_value:.2f}")
print(f"Total Return: {total_return_pct:.2f}%")

# 8. Plot the Equity Curve (Your Profit over Time)
import plotly.express as px
fig = px.line(backtest_df, x=backtest_df.index, y='Portfolio_Value', title='Backtest Performance: Equity Curve')
fig.show()
```

```python
%%writefile app.py
import streamlit as st
import yfinance as yf
import pandas as pd  # <--- This was the missing line!
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# --- 1. Sidebar Configuration ---
st.set_page_config(layout="wide", page_title="Algo Trading Dashboard")
st.sidebar.header("User Input")

ticker = st.sidebar.text_input("Ticker Symbol", "AAPL")
# Default dates
start_date = st.sidebar.date_input("Start Date", value=pd.to_datetime("2023-01-01"))
end_date = st.sidebar.date_input("End Date", value=pd.to_datetime("today"))

# --- 2. Data Loading Function ---
@st.cache_data
def load_data(symbol, start, end):
    try:
        data = yf.download(symbol, start=start, end=end)
        # Handle cases where MultiIndex columns might occur with new yfinance versions
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = data.columns.get_level_values(0)
        return data
    except Exception as e:
        st.error(f"Error loading data: {e}")
        return pd.DataFrame()

# --- 3. Main Logic ---
if st.sidebar.button("Analyze Stock"):
    # A. Fetch Data
    st.write(f"Fetching data for **{ticker}**...")
    data = load_data(ticker, start_date, end_date)

    if not data.empty:
        # B. Calculate Indicators
        # Ensure we have enough data
        if len(data) > 20:
            data['RSI'] = data.ta.rsi(length=14)
            bbands = data.ta.bbands(length=20, std=2)
            data = data.join(bbands)
            data.dropna(inplace=True)

            # C. Backtest Strategy (RSI Reversion)
            data['Signal'] = 0
            data.loc[data['RSI'] < 30, 'Signal'] = 1  # Buy
            data.loc[data['RSI'] > 70, 'Signal'] = -1 # Sell
            data['Position'] = data['Signal'].replace(to_replace=0, method='ffill')

            data['Market_Return'] = data['Close'].pct_change()
            data['Strategy_Return'] = data['Position'].shift(1) * data['Market_Return']

            initial_capital = 10000
            data['Portfolio_Value'] = initial_capital * (1 + data['Strategy_Return']).cumprod()

            total_return = ((data['Portfolio_Value'].iloc[-1] / initial_capital) - 1) * 100

            # --- 4. Dashboard Layout ---

            # Header Metrics
            col1, col2, col3 = st.columns(3)
            current_price = data['Close'].iloc[-1]
            final_value = data['Portfolio_Value'].iloc[-1]

            # Formatting checks
            col1.metric("Current Price", f"${current_price:.2f}")
            col2.metric("Total Return", f"{total_return:.2f}%")
            col3.metric("Final Portfolio Value", f"${final_value:.2f}")

            # Plot 1: Price + Bollinger Bands + RSI
            fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1,
                                subplot_titles=('Price & Bollinger Bands', 'RSI Indicator'), row_width=[0.3, 0.7])

            # Candlestick
            fig.add_trace(go.Candlestick(x=data.index, open=data['Open'], high=data['High'],
                                         low=data['Low'], close=data['Close'], name='OHLC'), row=1, col=1)

            # Bollinger Bands
            fig.add_trace(go.Scatter(x=data.index, y=data['BBU_20_2.0'], line=dict(color='gray', width=1), name='Upper BB'), row=1, col=1)
            fig.add_trace(go.Scatter(x=data.index, y=data['BBL_20_2.0'], line=dict(color='gray', width=1), name='Lower BB'), row=1, col=1)

            # RSI
            fig.add_trace(go.Scatter(x=data.index, y=data['RSI'], line=dict(color='purple', width=2), name='RSI'), row=2, col=1)
            fig.add_hline(y=70, line_dash="dot", row=2, col=1)
            fig.add_hline(y=30, line_dash="dot", row=2, col=1)

            fig.update_layout(height=800, xaxis_rangeslider_visible=False)
            st.plotly_chart(fig, use_container_width=True)

            # Plot 2: Backtest Equity Curve
            st.subheader("Strategy Performance (Equity Curve)")
            st.line_chart(data['Portfolio_Value'])
        else:
            st.warning("Not enough data to calculate indicators. Try a longer date range.")
    else:
        st.error("No data found for this ticker.")

else:
    st.info("👈 Enter a stock ticker in the sidebar and click 'Analyze Stock' to start.")
```
